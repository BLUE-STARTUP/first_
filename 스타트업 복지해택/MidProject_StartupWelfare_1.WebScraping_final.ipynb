### 회사 주소 스크래핑-로그인 안하는 버전

```python
page_num = 1
job_search_url = "https://www.rocketpunch.com/companies?funding_volume=10000000000-" + '&page=' + f'{page_num}'
driver.get(job_search_url)
```

```python
last_page = driver.find_element_by_xpath('//*[@id="search-results"]/div[2]/div/div[1]/a[6]')
last_page_num = last_page.get_attribute('data-query-add')
last_page_num = int(last_page_num.replace('page=',''))

url_arr = []
url_number = 1
url_count = 0

while(True):
    try:
        time.sleep(1) #컴퓨터 속도에 따라 sleep 안의 수 변경해주세요
        url = driver.find_element_by_xpath(f'//*[@id="company-list"]/div[{url_number}]/div[2]/a')
        #css = driver.find_element_by_css_selector(f'#company-list > div:nth-child({url_number}) > div.content > a')
        url_arr.append(url.get_attribute('href'))
        print(f"{url_count+1}개의 link를 가져왔습니다. "+ url.get_attribute('href'))
        
        url_number += 2
        url_count += 1
    except NoSuchElementException as Nse :
        if url_number == 7 :
            url_number += 2
            continue
        if page_num <= last_page_num :
            url_number = 1
            page_num += 1
            job_search_url = "https://www.rocketpunch.com/companies?funding_volume=10000000000-" + '&page=' + f'{page_num}'
            driver.get(job_search_url)
            print(f'다음 페이지로 넘어갑니다. 현재 페이지 : {page_num}')
            continue
        else:
            print('스크래핑이 완료되었습니다.')
            break
        
```

### 복지정보 스크래핑

```python
#채용정보 사이트 URL 만들기
url_jobs = []

for url in url_arr:
    
    url_jobs.append(url+'/jobs')
```

```python
df = pd.DataFrame()
```

```python
#복지정보 스크래핑
for num, url in enumerate(url_jobs):
    try :
        time.sleep(1) #컴퓨터 속도에 따라 sleep 안의 수 변경해주세요
        driver.get(url)
        print(str(num+1)+'번째를 가져왔습니다.'+url)

        welfare = driver.find_element_by_xpath('//*[@id="page-benefits"]/div')
        title = driver.find_element_by_xpath('//*[@id="company-name"]/h1')
        
        title_text = title.text
        welfare_text = welfare.text
        welfare_text_list = welfare_text.split('\n')

        welfare_category_list = []
        welfare_contents_list = []
        
        #복지 종류와 내용을 분류
        for num, x in enumerate(welfare_text_list) :
                if (num ==0) or (num % 2 ==0) :
                    welfare_category_list.append(welfare_text_list[num])
                else :
                    welfare_contents_list.append(welfare_text_list[num])

        welfare_text_dict = dict(zip(welfare_category_list,welfare_contents_list))
        df_welfare = pd.DataFrame([welfare_text_dict], index=[title_text])
        df = df.append(df_welfare)
        
    except NoSuchElementException as nse :
        print("위 기업은 복지 정보를 가지고 있지 않습니다.")

        continue
```
