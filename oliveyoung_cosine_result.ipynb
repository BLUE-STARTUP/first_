```python
import warnings
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer  # 피체 백터화
from sklearn.metrics.pairwise import cosine_similarity # 코사인 유사도
from konlpy.tag import Twitter
twitter = Twitter()
warnings.filterwarnings("ignore")
```

```python
df_new = pd.read_excel('oliveyoung_df.xlsx')
```

# review 코사인 유사도

```python
# review 명사값만 추출
review_list = []
for i in range(len(df_new['review'])):
    nouns = twitter.nouns(df_new['review'][i])
    nouns = ' '.join(nouns)
    review_list.append(nouns)
```

```python
df_cosine=pd.DataFrame()
df_cosine['review_noun'] = review_list
```

```python
#리뷰 데이터 코사인 유사도
count_vect_category = CountVectorizer(min_df=0, ngram_range=(1,2))
review = count_vect_category.fit_transform(df_cosine['review_noun']) 
review_cosine = cosine_similarity(review,review) 
```

# 결과물2 input(피부 고민) text 를 리뷰데이터에 포함시켜 코사인 유사도가 비슷한 제품 추천

```python
oliveyoung = pd.read_excel('oliveyoung_df.xlsx')
```

```python
place_simi_co = review_cosine



# 아래 place_simi_co_sorted_ind 는 그냥 바로 사용하면 됩니다.
sorted_ind = place_simi_co.argsort()[:, ::-1] 


# 최종 구현 함수
def find_simi_review(df, sorted_ind, place_name, top_n=10):
    nouns = twitter.nouns(place_name)
    nouns = ' '.join(nouns)
    df.loc[len(df)] = ['','','',nouns,'','','','','','','','']
    
    # 리뷰 텍스트 데이터 간의 텍스트 피쳐 벡터라이징
    count_vect_review = CountVectorizer(min_df=2, ngram_range=(1,2))
    place_review = count_vect_review.fit_transform(df['review']) 

    # 리뷰 텍스트 간의 코사인 유사도 따지기
    place_simi_review = cosine_similarity(place_review, place_review)
    sorted_ind = place_simi_review.argsort()[:, ::-1]
    
    
    similar_indexes = sorted_ind[len(df)-1, :(top_n+1)]
#     return df.iloc[similar_indexes]
    df_print = df.iloc[similar_indexes]

    for i in similar_indexes:
        if i==623:
            pass
        else:
            print("------------------------------------------------------------------------------------------------------------")
            print("\n")
            print("✔️제품명 : " , df_print["name"][i])
            print("✔️브랜드 : " , df_print["brand"][i])
            print("✔️가격 : " , df_print["price"][i])
            print("\n바로 구매해 보세요 \n ☞ ", df_print["url"][i]) 
            print(" (리뷰수 :",df_print["product_total_review"][i],"건)")
            print("\n")
    df.drop(df.tail(1).index,inplace=True)
```

```python
find_simi_review(oliveyoung,sorted_ind, input(), 4)
```
